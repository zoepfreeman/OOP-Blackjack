### Rules:

`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>
**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust. 

```python
import random
```

```python
def new_deck():
    card_deck=[]
    suits=['c','d','h','s']
    for suit in suits:
        for i in range(1,14):
            card_deck.append(str(i)+suit)
    return card_deck

def read_cards(cards):
    card_list=[]
    card_string=''
    if type(cards)==str:
        cards=[cards]
    for card in cards:
        value=int(card[:-1])
        if value==1:
            value='Ace'
        elif value==11:
            value='Jack'
        elif value==12:
            value='Queen'
        elif value==13:
            value='King'
        if card[-1]=='c':
            card_list.append(f'{value} of Clubs')
        elif card[-1]=='d':
            card_list.append(f'{value} of Diamonds')
        elif card[-1]=='h':
            card_list.append(f'{value} of Hearts')
        elif card[-1]=='s':
            card_list.append(f'{value} of Spades')
    if len(card_list)>2:
        for i in range(len(card_list)-1):
            card_string+=card_list[i]+', '
        card_string+='and '+card_list[-1]
    elif len(card_list)==1:
        card_string+=card_list[0]
        return card_string
    else:
        card_string+=card_list[0]+', '
        card_string+=card_list[-1]
    return card_string
```

```python
class Blackjack:
    def __init__(self):
        self.card_deck=new_deck()
        self.player_cards=random.choices(self.card_deck,k=2)
        for i in range(2):
            self.card_deck.remove(self.player_cards[i])
        self.dealer_cards=random.choices(self.card_deck,k=2)
        for i in range(2):
            self.card_deck.remove(self.dealer_cards[i])
            
    def player_total(self):
        player_cards=self.player_cards
        player_total=0
        for card in player_cards:
            value=int(card[:-1])
            player_total+=value
        return player_total
    
    def dealer_total(self):
        dealer_cards=self.dealer_cards
        dealer_total=0
        for card in dealer_cards:
            value=int(card[:-1])
            dealer_total+=value
        return dealer_total
    
    def hit_player(self):
        new_card=random.choice(self.card_deck)
        self.card_deck.remove(new_card)
        self.player_cards.append(new_card)
        
    def hit_dealer(self):
        new_card=random.choice(self.card_deck)
        self.card_deck.remove(new_card)
        self.dealer_cards.append(new_card)
```

```python
def play_blackjack():
    game=Blackjack()
    print('Your cards: ',read_cards(game.player_cards))
    print(f"One of the dealer's cards is the {read_cards(random.choice(game.dealer_cards))}.")
    if game.player_total()==21:
        print('Blackjack! You win!!!')
    while game.player_total()<21:
        answer=input(f'Your total is {game.player_total()}. Would you like to hit or stand? h/s: ')
        if answer=='h':
            game.hit_player()
            print(read_cards(game.player_cards))
            print(f'New card: {read_cards(game.player_cards[-1])}')
        if answer=='s':
            break
    if game.player_total()>21:
        print(f'Your total is {game.player_total()}. Bust! You lose:(')
    else:
        while game.dealer_total()<17:
            game.hit_dealer()
        print()
        print()
        print(f"Your cards: {read_cards(game.player_cards)}")
        print(f"Your total: {game.player_total()}")
        print()
        if game.dealer_total()>=game.player_total() and game.dealer_total()<22:
            print(f"Dealer's cards: {read_cards(game.dealer_cards)}")
            print(f"Dealer's total: {game.dealer_total()}")
            print()
            print('You lose:(')
        else:
            print(f"Dealer's cards: {read_cards(game.dealer_cards)}")
            print(f"Dealer's total: {game.dealer_total()}")
            print()
            print('You win!!!')

            

play_blackjack()
```
